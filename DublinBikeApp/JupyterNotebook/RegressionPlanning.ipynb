```python
import pandas as pd
import numpy as np
import requests
import json
import gviz_api
from pandas.io.json import json_normalize
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_columns', None)  
```

```python
chart1 = requests.get("http://localhost:5000/chart/1")
```

```python
myJson = json.loads(chart1.text)
```

```python
mon = json_normalize(myJson['mon']).T
tue = json_normalize(myJson['tue']).T
wed = json_normalize(myJson['wed']).T
thu = json_normalize(myJson['thu']).T
fri = json_normalize(myJson['fri']).T
sat = json_normalize(myJson['sat']).T
sun = json_normalize(myJson['sun']).T
```

```python
df = pd.concat([mon, tue, wed, thu, fri, sat, sun], axis=1)
```

```python
df.columns = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']
```

```python
df.head(5)
```

```python
df.reset_index(inplace=True)
```

```python
df.head(5)
```

```python
df = df.applymap(float)
```

```python
df.head(5)
```

```python
df = df.sort_values(by=['index'])
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df.drop(['index'], axis=1)
```

```python
# Create chart for all days
fig, ax = plt.subplots(figsize=(12, 8))

dim = np.arange(1, 24)

# Use ax for both
df['mon'].plot(x=df.index, y='mon', ax=ax, label="Monday")
df['tue'].plot(x=df.index, y='tue', ax=ax, label="Tuesday")
df['wed'].plot(x=df.index, y='tue', ax=ax, label="Wednesday")
df['thu'].plot(x=df.index, y='wed', ax=ax, label="Thursday")
df['fri'].plot(x=df.index, y='thu', ax=ax, label="Friday")
df['sat'].plot(x=df.index, y='fri', ax=ax, label="Saturday")
df['sun'].plot(x=df.index, y='sat', ax=ax, label="Sunday")

ax.set_title("Hourly Average Available Bikes by Day of Week")
ax.set_xlabel('Hours')
ax.set_ylabel('Available Bikes')
ax.set_xticks(np.arange(0, 24))
ax.set_xticklabels([str(i) for i in range(1, 25)])

# Shrink current axis by 20%
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.85))

plt.show()
```
